```python
import csv

filepath = "data/heatmapdata_expset2_aggro_paladin.csv"

data_dict = csv.DictReader(open(filepath), delimiter=';')

fitness_value_set = set()
average_axis_value_set = set()
average_axis_value_list = []
variance_axis_value_set = set()
variance_axis_value_list = []

for row in data_dict:
    fitness_value_set.add(int(row['fitness']))
    average_axis_value_set.add(float(row['average']))
    average_axis_value_list.append(float(row['average']))
    variance_axis_value_set.add(float(row['variance']))
    variance_axis_value_list.append(float(row['variance']))

#print(sorted(fitness_value_set))
#print(sorted(average_axis_value_set))
#print(sorted(variance_axis_value_set))
```

```python
import math

number_of_bins = 5

average_bins = []
average_bins_lens = []
average_set_sorted = sorted(average_axis_value_list)
total_per_bin = math.ceil(len(average_set_sorted) / number_of_bins)

number_of_smaller_bins = number_of_bins - (len(average_set_sorted) % number_of_bins)
for i in range(0, number_of_bins):
    average_bins_lens.append(total_per_bin)
for i in range(0, number_of_smaller_bins):
    x = 0 - (i+1)
    average_bins_lens[x] -= 1

for j in range(0, len(average_bins_lens)):
    average_bins.append(average_set_sorted[sum(average_bins_lens[0:j]):(sum(average_bins_lens[0:j])+average_bins_lens[j])])

average_bins.reverse()
#print(average_bins_lens)
#print(sum(average_bins_lens))
print(average_bins)

print("################################################################")

variance_bins = []
variance_set_sorted = sorted(variance_axis_value_list)
    
for j in range(0, len(variance_bins_lens)):
    variance_bins.append(variance_set_sorted[sum(average_bins_lens[0:j]):(sum(average_bins_lens[0:j])+average_bins_lens[j])])

variance_bins

#print(variance_bins_lens)
#print(sum(variance_bins_lens))
print(variance_bins)

```

```python
def find_point_location(point, avg_bins, var_bins):
    x = -1
    y = -1
    
    fitness = int(point['fitness'])
    average = float(point['average'])
    variance = float(point['variance'])
    
    for i in range(0, len(avg_bins)):
        if average in avg_bins[i]:
            x = i
            break

    for i in range(0, len(var_bins)):
        if variance in var_bins[i]:
            y = i
            break
    
    return x, y, fitness
```

```python
import csv

filepath = "data/heatmapdata_expset2_aggro_paladin.csv"

data_dict = csv.DictReader(open(filepath), delimiter=';')

fitness_total_data = []
point_count_data = []
result_data = []
for x in range(0, number_of_bins):
    fitness_total_data.append([])
    point_count_data.append([])
    result_data.append([])
    for y in range(0, number_of_bins):
        fitness_total_data[-1].append(0)
        point_count_data[-1].append(0)
        result_data[-1].append(0)

for row in data_dict:
    x, y, fit = find_point_location(row, average_bins, variance_bins)
    
    fitness_total_data[x][y] += fit
    point_count_data[x][y] += 1

print(fitness_total_data)
print(point_count_data)

min_value = None
max_value = None

for i in range(0, len(result_data)):
    for j in range(0, len(result_data[i])):
        if float(point_count_data[i][j]) == 0:
            result_data[i][j] = -math.inf
        else:
            result_data[i][j] = float(fitness_total_data[i][j]) / float(point_count_data[i][j])
            if min_value == None or result_data[i][j] < min_value:
                min_value = result_data[i][j]
            if max_value == None or result_data[i][j] > max_value:
                max_value = result_data[i][j]

for i in range(0, len(result_data)):
    for j in range(0, len(result_data[i])):
        if result_data[i][j] != -math.inf:
            result_data[i][j] = float(result_data[i][j] - min_value) / float(max_value - min_value)
        else:
            pass
            #result_data[i][j] = 0.0

print(min_value)
print(max_value)
print(result_data)
```

```python
import seaborn as sns
#import numpy as np
import pandas as pd

sns.set()
#heatmap_data = np.array(result_data)
heatmap_data = pd.DataFrame.from_items([('high-avg', result_data[0]), ('highmid-avg', result_data[1]), ('mid-avg', result_data[2]), ('lowmid-avg', result_data[3]), ('low-avg', result_data[4])], orient='index', columns=['low-var', 'lowmid-var', 'mid-var', 'highmid-var', 'high-var'])
print(heatmap_data)
ax = sns.heatmap(heatmap_data, center=0.5, cmap="YlGnBu")
```

```python
import csv

def load_data(filepath):
    data_dict = csv.DictReader(open(filepath), delimiter=';')

    data_list = []

    for row in data_dict:
        data_list.append({'fitness': int(row['fitness']), 'average': float(row['average']), 'variance': float(row['variance']), 'deck_id': row['deck_id']})
    
    return data_list
```

```python
import seaborn as sns
import pandas as pd

filepath = "data/heatmapdata_expset2_aggro_paladin.csv"
experiment = "ExpSet2 Paladin Aggro"

data_list = load_data(filepath)

sns.set()
sp_data = pd.DataFrame(data_list)
ax = sns.scatterplot(x="average", y="variance", hue="fitness", size="fitness", sizes=(10, 100), data=sp_data).set_title(experiment)
```

```python
import seaborn as sns
import pandas as pd
import csv

path = "Experiments/ExpSet1/Control/Paladin/"
temp = path.split("/")
filepath = "data/heatmapdata_" + temp[1].lower() + "_" + temp[2] + "_" + temp[3] + ".csv" #expset1_control_warlock.csv"
experiment = "Sunwalker " + temp[1] + " " + temp[2] + " " + temp[3] #"ExpSet1 Control Warlock"

data_list = load_data(filepath)

#########

individuals_log_file = csv.DictReader(open(path + "individual_log.csv"))

deck_dict = {}

for row in individuals_log_file:
    deck_dict[row["Individual"]] = row["Deck"].split('*')

items_to_remove = []
print(len(data_list))
    
for entry in data_list:
    if "Sunwalker" not in deck_dict[entry['deck_id']]:
        items_to_remove.append(entry)

for item in items_to_remove:
    data_list.remove(item)

print(len(data_list))

#########

sns.set()
sp_data = pd.DataFrame(data_list)
ax = sns.scatterplot(x="average", y="variance", hue="fitness", size="fitness", sizes=(10, 100), data=sp_data).set_title(experiment)
```
